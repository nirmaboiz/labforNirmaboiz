```python
import pandas as pd
```

```python
airData = pd.read_csv("airquality.csv")
```

```python
airData.isnull().sum()
```

```python
airData.head()
```

```python
#Calculating the Mean of Ozone column
mean_ozone = airData['Ozone'].mean()
```

```python
#Substituting the null value in Ozone column with the mean of Ozone
airData['Ozone'].fillna(mean_ozone,inplace=True)
# Column Ozone cleaned
```

```python
airData.isnull().sum()
```

```python
# Cleaned Solar.R column by replacing null with the mean of the column
airData['Solar.R'].fillna(airData['Solar.R'].mean(),inplace=True)
```

```python
airData.isnull().sum()
```

```python
airData['NewSolar'] = [True if x1>200 else False for x1 in airData['Solar.R']] 
```

```python
airData.head()
```

```python
airData.isnull().sum()
```

```python
airData['NewTemp'] = ['Hot' if x1>60 else 'cold' for x1 in airData['Temp']] 
```

```python
airData.head()
```

```python
vec=[0,50,100,150,200,250,300,350]
airData['Solar.R'] = pd.cut(airData['Solar.R'],bins=vec)
```

```python
airData.head(20)
```

```python
ozone_range=[0,10,20,30,40,50]
airData['Ozone'] = pd.cut(airData['Ozone'],bins=ozone_range)
```

```python
airData.head(20)
```

```python
wind_range=[0,5,10,15,20,25]
airData['Wind'] = pd.cut(airData['Wind'],bins=wind_range)
```

```python
airData.head(20)
```

```python
airData.loc[airData['Month']==5,"Month"] = "May"
airData.loc[airData['Month']==6,"Month"] = "June"
```
